### 3. Regression Modelling

```python
X = df[['VAR3','VAR2','Unit1','Unit2','Unit3','Unit4','Unit5','Unit6','Unit7','Unit8','Unit9','Unit10','Unit11','Unit12','Time_of_service','Age', 'Education', 'Relationship_Status', 'Time_since_promotion', 'growth_rate', 'Pay_Scale', 'Post_Level','Work_Life_balance']].values
Y = df['Attrition_rate'].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
# from sklearn.ensemble import RandomForestRegressor
# reg = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# reg.fit(X_train, y_train);
reg = linear_model.Lasso()
# reg = LinearRegression()  
# reg=linear_model.Ridge(alpha=.4)
reg.fit(X_train, y_train)
```

```python
# reg.coef_
```

```python
coef = pd.DataFrame()
coef['Coefficient']=reg.coef_
a=['VAR3','VAR2','Unit1','Unit2','Unit3','Unit4','Unit5','Unit6','Unit7','Unit8','Unit9','Unit10','Unit11','Unit12','Time_of_service','Age', 'Education', 'Relationship_Status', 'Time_since_promotion', 'growth_rate',  'Pay_Scale','Post_Level', 'Work_Life_balance']
coef['Variable']=np.array(a)
coef
```

```python
y_pred = reg.predict(X_test)
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df2 = df1.head(25)
df2
```
